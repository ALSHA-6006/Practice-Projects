```python
#Here i am using label encoder to change all categorical values to numerical values
import sklearn
from sklearn.preprocessing import LabelEncoder
```

```python
import warnings
warnings.filterwarnings('ignore')
le=LabelEncoder()
list=['Name','Sex','Ticket','Cabin','Embarked']
for val in list:
    df[val]=le.fit_transform(df[val].astype(str))
```

```python
df
```

```python
#removing outliers
#"abs" stands for absolute it will convert all -ve values to +ve values 
import scipy
import numpy as np
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
#here we are using a threshold, because we only need values beyond 3
threshold=3
np.where(z>3)
```

```python
z.iloc[6][10]
```

```python
#all(axis=1) means to check all the columns having values <3.
import warnings
warnings.filterwarnings('ignore')
df_new=df[(z<3).all(axis=1)]
df_new
```

```python
df_new.shape
```

so, above code removed around 105 rows from the dataset ie, outliers have been removed

```python
x=df_new.iloc[:,:-1]
x
```

```python
y=df_new.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
pred=lg.predict(x_test)
pred
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
